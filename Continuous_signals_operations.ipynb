# Continuous signal operations

Continuous signals are fundamental in various fields, including signal processing, communications, and engineering. This document explores key operations that can be performed on continuous signals.

# 1. Superposition

# Definition

Superposition is a principle that states when two or more signals are combined, the resulting signal is the algebraic sum of the individual signals at each point in time.

# Mathematical Representation

For two continuous signals f(t) and g(t), the superposition operation is defined as:

$h(t)= f(t) + g(t)$

# Properties

- Linear operation

- Preserves the characteristics of individual signals

- Fundamental in linear system analysis

# Examples

- Combining audio signals

- Analyzing interference in wave systems

- Electronic circuit analysis

#Code example

```python
import sympy as sym

def superposition (x1, x2):

  x = x1 + x2
  return x

t = sym.symbols("t", real = True) #temporal variable

x1 = 3*sym.cos(3*t) #signal 1
x2 = 5*sym.cos(t) #signal 2

x = superposition(x1, x2) #superposition operation

sym.plot(x, (t, 0, 2*sym.pi), ylim = (-10, 10), xlabel = "t", ylabel = "x(t)");
```

#2. Temporal displacement

#Definition

Temporal displacement (or time shifting) involves moving a signal along the time axis without changing its fundamental shape or amplitude.

#Mathematical Representation

For a signal f(t) displaced by τ units:

- Right shift: f(t - τ)

- left shift: f(t + τ)

#Key Characteristics

- Preserves signal shape

- Changes the time origin

- Useful in synchronization and delay analysis

#Applications

- Audio and signal synchronization

- Radar and communication systems

- Time-domain signal analysis



#Code example

```python
rect= sym.Heaviside(t + 1/2) - sym.Heaviside(t - 1/2) #creting a rectangular pulse
sym.plot(rect, (t, -2*sym.pi, 2*sym.pi), ylim = (0, 2), xlabel = "t", ylabel = "x(t)");
```

#Example 2

```python
rect2 = rect.subs(t, t-1/2) + 2/3*rect.subs(t, t-3/2) + 1/3*rect.subs(t, t-5/2) #creating a shifted rectangular pulse
sym.plot(rect2, (t, -1.5, 4), ylim = (0, 1.5), xlabel = "t", ylabel = "x(t)");
```

# 3. Scaling Displacement

# Definition
Scaling displacement involves changing the amplitude or duration of a signal while maintaining its overall shape.

#Types of Scaling

#1. Amplitude Scaling

- Multiplies the signal by a constant factor

- Increases or decreases signal magnitude

#2 Time Scaling

- Compresses or stretches the signal along the time axis

- Preserves relative signal characteristics

#Mathematical Representation

$g(t) = f(k * t)$

- k > 1: Signal compression
- 0 < k < 1: Signal stretching

#Significance

- Signal normalization
- Frequency domain transformations
- Dynamic range adjustment

#Code example

```python
a = sym.symbols("a", real= True) #creating scaling parameter
y = rect2.subs(t, a*t) #scaling rect2
px = sym.plot(rect2, (t, -1.5, 4), ylim = (0, 1.5), xlabel = "t", ylabel = "x(t)");
py = sym.plot(y.subs(a, 2), (t, -1.5, 4), ylim = (0, 1.5), xlabel = "t", ylabel = "x(t)"); #a=2 for compression
px.extend(py) #2 graphs in the same figure
px.show;
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Original sinusoidal signal
t_continuous = np.linspace(0, 1, 1000)
continuous_signal = np.sin(2 * np.pi * 10 * t_continuous)

# Sampling at different frequencies
fs1 = 20  # Sufficient
fs2 = 15  # Undersampling

t1 = np.linspace(0, 1, int(fs1))
t2 = np.linspace(0, 1, int(fs2))

samples1 = np.sin(2 * np.pi * 10 * t1)
samples2 = np.sin(2 * np.pi * 10 * t2)

plt.figure(figsize=(10, 6))
plt.plot(t_continuous, continuous_signal, label='Continuous Signal')
plt.scatter(t1, samples1, color='red', label=f'Sampling {fs1} Hz')
plt.scatter(t2, samples2, color='green', label=f'Undersampling {fs2} Hz')
plt.legend()
plt.title('Sampling Comparison')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.gridspec import GridSpec

# Configuración general
fig = plt.figure(figsize=(10, 8))
gs = GridSpec(3, 1, height_ratios=[1, 0.5, 1])

# Señal analógica original
ax1 = fig.add_subplot(gs[0])
t = np.linspace(0, 1, 1000)
f = 5  # Frecuencia de la señal
signal = np.sin(2 * np.pi * f * t)
ax1.plot(t, signal, 'b-', linewidth=2)
ax1.set_title('Señal Continua en el Dominio de la Frecuencia $X(f)$', pad=20)
ax1.set_ylabel('Amplitud')
ax1.text(0.5, -1.5, r"$X(f) \in \mathbb{R}$ (Señal continua en frecuencia)",
         ha='center', fontsize=12, bbox=dict(facecolor='white', alpha=0.8))

# Proceso de muestreo (representado por H)
ax2 = fig.add_subplot(gs[1])
ax2.text(0.5, 0.5, r"Proceso de Muestreo: $H = \Delta t \cdot \sum_{n=-\infty}^{\infty} \delta(t - nT_s)$",
         ha='center', va='center', fontsize=14)
ax2.axis('off')

# Señal muestreada
ax3 = fig.add_subplot(gs[2])
Ts = 0.05  # Periodo de muestreo
t_samples = np.arange(0, 1, Ts)
samples = np.sin(2 * np.pi * f * t_samples)
ax3.stem(t_samples, samples, linefmt='r-', markerfmt='ro', basefmt=" ")
ax3.plot(t, signal, 'b--', alpha=0.3)
ax3.set_title('Señal Muestreada en el Dominio del Tiempo $tS$', pad=20)
ax3.set_xlabel('Tiempo (t)')
ax3.set_ylabel('Amplitud')
ax3.text(0.5, -1.5, r"$t \in \mathbb{R}$, $T_s$ = Periodo de muestreo",
         ha='center', fontsize=12, bbox=dict(facecolor='white', alpha=0.8))

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import FancyArrowPatch

# Configuración del gráfico
plt.figure(figsize=(12, 6))
plt.style.use('seaborn-v0_8-whitegrid')  # Estilo actualizado

# Parámetros de la señal
fs = 1000  # Frecuencia de muestreo (Hz)
Ts = 1/fs  # Periodo de muestreo (s)
f = 100    # Frecuencia de la señal (Hz)
t_total = 0.03  # Tiempo total mostrado (s)

# Generar señal continua
t_cont = np.linspace(0, t_total, 5000)
signal_cont = 80 * np.sin(2 * np.pi * f * t_cont)

# Generar señal muestreada
t_samples = np.arange(0, t_total, Ts)
signal_samples = 80 * np.sin(2 * np.pi * f * t_samples)

# Dibujar señal continua
plt.plot(t_cont, signal_cont, 'b-', alpha=0.5, linewidth=1, label='Señal analógica')

# Dibujar muestras
markerline, stemlines, baseline = plt.stem(t_samples, signal_samples,
                                         linefmt='r-', markerfmt='ro', basefmt=' ',
                                         label='Muestras')
plt.setp(stemlines, 'linewidth', 1)
plt.setp(markerline, 'markersize', 6)

# Configuración de ejes
plt.title('Señal Muestreada con Indicación de $T_s$', pad=20, fontsize=14)
plt.xlabel('Tiempo (s)', fontsize=12, labelpad=10)
plt.ylabel('Amplitud [V]', fontsize=12, labelpad=10)
plt.ylim(-105, 105)
plt.grid(True, linestyle='--', alpha=0.7)

# Añadir indicación de Ts (entre las muestras 5 y 6 como ejemplo)
sample1, sample2 = 5, 6
mid_point = (t_samples[sample1] + t_samples[sample2])/2

# Flecha indicando Ts
arrow = FancyArrowPatch((t_samples[sample1], -20), (t_samples[sample2], -20),
                        arrowstyle='<->', mutation_scale=15, color='green', linewidth=2)
plt.gca().add_patch(arrow)

# Texto indicando Ts
plt.text(mid_point, -30, r'$T_s = {:.4f}$ s'.format(Ts),
         ha='center', va='top', fontsize=12, bbox=dict(facecolor='white', alpha=0.8))

# Líneas verticales punteadas para guía
plt.plot([t_samples[sample1], t_samples[sample1]], [0, signal_samples[sample1]],
         'k--', alpha=0.5, linewidth=0.8)
plt.plot([t_samples[sample2], t_samples[sample2]], [0, signal_samples[sample2]],
         'k--', alpha=0.5, linewidth=0.8)

# Leyenda
plt.legend(loc='upper right', fontsize=10)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
from matplotlib.path import Path

# Configurar la figura
fig, ax = plt.subplots(figsize=(10, 6))

# Ocultar ejes
ax.axis('off')
ax.set_xlim(0, 10)
ax.set_ylim(0, 6)
ax.set_aspect('equal')
ax.set_title('Circuito Sample-and-Hold para Muestreo de Señales', pad=20, fontsize=14)

# Dibujar componentes

# 1. Señal de entrada
ax.plot([1, 3], [4, 4], 'b-', lw=2)
ax.text(0.8, 4.1, '$X_c(t)$', fontsize=12)

# 2. Switch (MOSFET)
# Símbolo del transistor
vertices = [
    (3, 4), (4, 4),  # Source
    (4, 4), (4, 3),   # Gate
    (4, 4), (5, 4)    # Drain
]
codes = [
    Path.MOVETO, Path.LINETO,
    Path.MOVETO, Path.LINETO,
    Path.MOVETO, Path.LINETO
]
path = Path(vertices, codes)
patch = patches.PathPatch(path, facecolor='none', lw=2)
ax.add_patch(patch)
ax.text(4.2, 2.8, 'Switch\n(MOSFET)', fontsize=10, ha='center')

# 3. Señal de control (clock)
ax.plot([4, 4], [3, 2], 'g-', lw=1)
ax.plot([3.8, 4.2], [2, 2], 'g-', lw=1)
ax.text(4.3, 2, 'Señal de\ncontrol', fontsize=10)

# 4. Capacitor
cap_x = 6
cap_y = 4
ax.plot([5, cap_x], [4, cap_y], 'k-', lw=2)
ax.plot([cap_x, cap_x], [cap_y-0.5, cap_y+0.5], 'k-', lw=2)
ax.plot([cap_x+0.5, cap_x+0.5], [cap_y-0.5, cap_y+0.5], 'k-', lw=2)
ax.text(cap_x+0.8, cap_y, 'C', fontsize=12)

# 5. Amplificador operacional
triangle = patches.Polygon([[7, 3.5], [8, 4], [7, 4.5]], closed=True, fill=False, lw=2)
ax.add_patch(triangle)
ax.plot([8, 9], [4, 4], 'r-', lw=2)
ax.text(9.1, 4, '$X[nT_s]$', fontsize=12)

# 6. Tiempo de muestreo
ax.text(4, 5, 'Tiempo de muestreo $T_s$', fontsize=12, ha='center',
        bbox=dict(facecolor='white', alpha=0.8))

# 7. Etiquetas
ax.text(2, 5.5, 'Señal analógica de entrada', fontsize=10, color='blue')
ax.text(7, 5.5, 'Señal muestreada', fontsize=10, color='red')

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
Fs = 1000  # Sampling frequency (Hz)
Ts = 1/Fs  # Sampling period
f = 50     # Signal frequency (Hz)
t = np.linspace(0, 0.1, 1000)  # 100 ms

# Continuous signal
x_cont = np.sin(2*np.pi*f*t)

# Sampled signal
n = np.arange(0, 0.1, Ts)
x_samp = np.sin(2*np.pi*f*n)

# Plot with corrected stem() function
plt.figure(figsize=(10,4))
plt.plot(t, x_cont, 'b-', label='Continuous')
markerline, stemlines, baseline = plt.stem(n, x_samp, 'r-', label='Samples')
plt.setp(markerline, color='red', markersize=5)
plt.setp(stemlines, color='red', linewidth=1)
plt.xlabel('Time (s)'); plt.ylabel('Amplitude')
plt.legend(); plt.grid()
plt.title(f'Signal Sampling (Fs={Fs}Hz, f={f}Hz)')
plt.show()
```

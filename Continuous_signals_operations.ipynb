# Continuous signal operations

Continuous signals are fundamental in various fields, including signal processing, communications, and engineering. This document explores key operations that can be performed on continuous signals.

# 1. Superposition

# Definition

Superposition is a principle that states when two or more signals are combined, the resulting signal is the algebraic sum of the individual signals at each point in time.

# Mathematical Representation

For two continuous signals f(t) and g(t), the superposition operation is defined as:

$h(t)= f(t) + g(t)$

# Properties

- Linear operation

- Preserves the characteristics of individual signals

- Fundamental in linear system analysis

# Examples

- Combining audio signals

- Analyzing interference in wave systems

- Electronic circuit analysis

#Code example

```python
import sympy as sym

def superposition (x1, x2):

  x = x1 + x2
  return x

t = sym.symbols("t", real = True) #temporal variable

x1 = 3*sym.cos(3*t) #signal 1
x2 = 5*sym.cos(t) #signal 2

x = superposition(x1, x2) #superposition operation

sym.plot(x, (t, 0, 2*sym.pi), ylim = (-10, 10), xlabel = "t", ylabel = "x(t)");
```

#2. Temporal displacement

#Definition

Temporal displacement (or time shifting) involves moving a signal along the time axis without changing its fundamental shape or amplitude.

#Mathematical Representation

For a signal f(t) displaced by τ units:

- Right shift: f(t - τ)

- left shift: f(t + τ)

#Key Characteristics

- Preserves signal shape

- Changes the time origin

- Useful in synchronization and delay analysis

#Applications

- Audio and signal synchronization

- Radar and communication systems

- Time-domain signal analysis



#Code example

```python
rect= sym.Heaviside(t + 1/2) - sym.Heaviside(t - 1/2) #creting a rectangular pulse
sym.plot(rect, (t, -2*sym.pi, 2*sym.pi), ylim = (0, 2), xlabel = "t", ylabel = "x(t)");
```

#Example 2

```python
rect2 = rect.subs(t, t-1/2) + 2/3*rect.subs(t, t-3/2) + 1/3*rect.subs(t, t-5/2) #creating a shifted rectangular pulse
sym.plot(rect2, (t, -1.5, 4), ylim = (0, 1.5), xlabel = "t", ylabel = "x(t)");
```

# 3. Scaling Displacement

# Definition
Scaling displacement involves changing the amplitude or duration of a signal while maintaining its overall shape.

#Types of Scaling

#1. Amplitude Scaling

- Multiplies the signal by a constant factor

- Increases or decreases signal magnitude

#2 Time Scaling

- Compresses or stretches the signal along the time axis

- Preserves relative signal characteristics

#Mathematical Representation

$g(t) = f(k * t)$

- k > 1: Signal compression
- 0 < k < 1: Signal stretching

#Significance

- Signal normalization
- Frequency domain transformations
- Dynamic range adjustment

#Code example

```python
a = sym.symbols("a", real= True) #creating scaling parameter
y = rect2.subs(t, a*t) #scaling rect2
px = sym.plot(rect2, (t, -1.5, 4), ylim = (0, 1.5), xlabel = "t", ylabel = "x(t)");
py = sym.plot(y.subs(a, 2), (t, -1.5, 4), ylim = (0, 1.5), xlabel = "t", ylabel = "x(t)"); #a=2 for compression
px.extend(py) #2 graphs in the same figure
px.show;
```
